Create a Python class for Student

Given the following problem definition, decide on the class and instance attributes and methods associated with the student class.

- Each student will have their unique student ID, name, program, sex, and weight and height.
- A student may enroll in a course (ex CPE103)
- Create the appropriate methods to compute for the student's midterm and final grade (include class standing quiz, seatwork, assignment, and recitation.
- Refer to the grade computation formula of the college.
- A function should be available to view the courses enrolled and the final grade per course


Testing
Create 3 student objects with the following names "Ana", "Mench" and "John". Test the methods created from the problem description.

```python
class Student:
  def __init__(self, student_ID, name, program, sex, weight, height):
    self.student_ID = student_ID
    self.name = name
    self.program = program
    self.sex = sex
    self.weight = weight
    self.height = height
    self.courses = {}

  def enroll_course(self, course_code):

    if course_code not in self.courses:

     self.courses[course_code] = {'midterm': 0, 'final' : 0, 'project': 0,'assignment': 0,'seatwork': 0, 'quiz': 0, 'recitation': 0, 'boardwork': 0, }
     print(f"{self.name} has enrolled in {course_code}.")

    else:
     print(f"{self.name} has aldready enrolled in {course_code}")

  def update_grades(self, course_code, midterm = 0, final = 0, project = 0, assignment = 0, seatwork = 0, quiz = 0, recitation = 0, boardwork = 0):

    if course_code in self.courses:
            self.courses[course_code].update({'midterm': midterm,'final': final,'project': project,'quiz': quiz,'seatwork': seatwork,'assignment': assignment, 'recitation': recitation,'boardwork': boardwork})
    else:
            print(f"{self.name} is not enrolled in {course_code}")

  def compute_final_grade(self, course_code):
    if course_code in self.courses:
            grades = self.courses[course_code]
            final_grade = (grades['midterm'] * 0.2 + grades['final'] * 0.3 + grades['quiz'] * 0.1 + grades['seatwork'] * 0.1 + grades['assignment'] * 0.1 + grades['recitation'] * 0.1 + grades['boardwork'] * 0.1)
            return round(final_grade, 2)

    else:
            print(f"{self.name} is not enrolled in {course_code}")
            return None

  def view_courses(self):
        if self.courses:
            print(f"\nSchool ID: {self.student_ID}")
            print(f"Name: {self.name}")
            print(f"Program: {self.program}")
            print(f"Sex: {self.sex}")
            print(f"Weight: {self.weight} kg")
            print(f"Heigth: {self.height} cm")
            print(f"\nCourses enrolled by {self.name}:")
            for course, grades in self.courses.items():
                final_grade = self.compute_final_grade(course)
                print(f"{course}: Final Grade = {final_grade}")
        else:
            print(f"{self.name} is not enrolled in any courses")


ana = Student("2024086-E", "Ana", "Computer Engineering", "Female", 50,  160)
mench = Student("2024002-E", "Mench", "Electronic Engineering", "Female", 63, 171)
john = Student("2024014-E", "John", "Industrial Engineering", "Male", 65, 175)

print("***First Student***\n")

ana.enroll_course("CPE103")
midterm = float(input("Enter Midterm Exam Grade for Ana in Average (0-100): "))
final = float(input("Enter Final Exam Grade for Ana in Average (0-100): "))
quiz = float(input("Enter Quiz Grade for Ana in Average (0-100): "))
seatwork = float(input("Enter Seatwork Grade for Ana in Average (0-100): "))
assignment = float(input("Enter Assignment Grade for Ana in Average (0-100): "))
boardwork = float(input("Enter Boardwork Grade for Ana in Average (0-100): "))
recitation = float(input("Enter Recitation Grade for Ana in Average (0-100): "))
ana.update_grades("CPE103", midterm, final, 0, assignment, seatwork, quiz, recitation, boardwork)
ana.view_courses()

print("\n***Second Student***\n")

mench.enroll_course("MAT002")
midterm = float(input("Enter Midterm Exam Grade for Mench in Average (0-100): "))
final = float(input("Enter Final Exam Grade for Mench in Average (0-100): "))
quiz = float(input("Enter Quiz Grade for Mench in Average (0-100): "))
seatwork = float(input("Enter Seatwork Grade for Mench in Average (0-100): "))
assignment = float(input("Enter Assignment Grade for Mench in Average (0-100): "))
boardwork = float(input("Enter Boardwork Grade for Mench in Average (0-100): "))
recitation = float(input("Enter Recitation Grade for Mench in Average (0-100): "))
mench.update_grades("MAT002", midterm, final, 0, assignment, seatwork, quiz, recitation, boardwork)
mench.view_courses()

print("\n***Third Student*** \n")

john.enroll_course("PHY001")
midterm = float(input("Enter Midterm Exam Grade for John in Average (0-100): "))
final = float(input("Enter Final Exam Grade for John in Average (0-100): "))
quiz = float(input("Enter Quiz Grade for John in Average (0-100): "))
seatwork = float(input("Enter Seatwork Grade for John in Average (0-100): "))
assignment = float(input("Enter Assignment Grade for John in Average (0-100): "))
boardwork = float(input("Enter Boardwork Grade for John in Average (0-100): "))
recitation = float(input("Enter Recitation Grade for John in Average (0-100): "))
john.update_grades("PHY001", midterm, final, 0, assignment, seatwork, quiz, recitation, boardwork)
john.view_courses()












```
